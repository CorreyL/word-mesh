# Example 1: Trump v Clinton

Open debate_transcript.txt

```python
with open('trump_hillary_debate.txt') as file:
    text = file.read()
```

Parse and filter the transcript

```python
transcript = text.split('\n\n')
```

```python
filtered_transcript = []
for para in transcript:
    if para.startswith('Trump:'):
        filtered_transcript.append((0,para[7:]))
    elif para.startswith('Clinton:'):
        filtered_transcript.append((1,para[9:]))
```

Import the module

```python
import sys, os
```

```python
package_path = os.path.dirname(os.getcwd())
sys.path.append(package_path)
```

```python
from wordmesh import Wordmesh, LabelledWordmesh
```

Create a LabelledWordmesh object, and extract the **40 most frequently used *adjectives*** for each label

```python
wm = LabelledWordmesh(filtered_transcript, num_keywords=40, extract_ngrams=False, 
                      keyword_extractor='tf', pos_filter=['ADJ'])
```

Cluster the extracted keywords on the canvas based on their **meaning**

```python
wm.set_clustering_criteria(by='meaning')
```

Color the keywords based on their label

```python
colormap = lambda x: (250,100,150) if (x==0) else (150,100,250)
wm.set_fontcolor(custom_colors = list(map(colormap, wm.labels)))
```

```python
wm.save_as_html(force_directed_animation=False, notebook_mode=True)
```
